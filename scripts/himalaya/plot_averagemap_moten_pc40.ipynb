```python
from nilearn.plotting import plot_surf
import matplotlib.pyplot as plt
from nilearn import datasets
from os.path import join
import os
```

```python
def plot_infl(gii_dir, align, key, pca, cmap):

    # align = os.path.basename(gii_dir)
    figure, axes = plt.subplots(3,2,subplot_kw={'projection': '3d'},figsize = (10,12))
    plot_surf(fsaverage['infl_left'],join(gii_dir,f'{key}_pca-{pca}_align-{align}_sub-mean_run-mean_hemi-lh.gii'), 
              title=f'{key} lateral lh', 
              hemi='left', view = 'lateral',cmap = cmap, axes=axes[0][0], colorbar=True, vmin=0, vmax=.1)
    plot_surf(fsaverage['infl_right'],join(gii_dir,f'{key}_pca-{pca}_align-{align}_sub-mean_run-mean_hemi-rh.gii'), 
              title=f'{key} lateral rh', 
              hemi='right', view = 'lateral', cmap = cmap, axes=axes[0][1], colorbar=True, vmin=0, vmax=.1)
    plot_surf(fsaverage['infl_left'],join(gii_dir,f'{key}_pca-{pca}_align-{align}_sub-mean_run-mean_hemi-lh.gii'), 
              title=f'{key} medial lh', 
              hemi='left', view = 'medial',cmap = cmap, axes=axes[1][0], colorbar=True, vmin=0, vmax=.1)
    plot_surf(fsaverage['infl_right'],join(gii_dir,f'{key}_pca-{pca}_align-{align}_sub-mean_run-mean_hemi-rh.gii'), 
              title=f'{key} medial rh', 
              hemi='right', view = 'medial', cmap = cmap, axes=axes[1][1], colorbar=True, vmin=0, vmax=.1)

    plot_surf(fsaverage['infl_left'],join(gii_dir,f'{key}_pca-{pca}_align-{align}_sub-mean_run-mean_hemi-lh.gii'), 
              title=f'{key} ventral lh', 
              hemi='left', view = 'ventral',cmap = cmap, axes=axes[2][0], colorbar=True, vmin=0, vmax=.1)
    plot_surf(fsaverage['infl_right'],join(gii_dir,f'{key}_pca-{pca}_align-{align}_sub-mean_run-mean_hemi-rh.gii'), 
              title=f'{key} ventral rh', 
              hemi='right', view = 'ventral', cmap = cmap, axes=axes[2][1], colorbar=True, vmin=0, vmax=.1)

    plt.show()
```
