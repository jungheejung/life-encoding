```python
from nilearn.plotting import plot_surf
import matplotlib.pyplot as plt
from nilearn import datasets
from os.path import join
```

```python
fsaverage = datasets.fetch_surf_fsaverage(mesh = 'fsaverage6')
```

```python
gii_dir = '/dartfs/rc/lab/D/DBIC/DBIC/f0042x1/life-encoding/results/himalaya/ha_common'
```

# r

```python
key = 'agents-r'
figure, axes = plt.subplots(1,2,subplot_kw={'projection': '3d'},figsize = (12,8))
plot_surf(fsaverage['infl_left'],join(gii_dir,f'{key}_align-ha_common_sub-mean_run-mean_hemi-lh.gii'), 
          title=f'{key} lh', 
          hemi='left', view = 'lateral',cmap = 'inferno',axes=axes[0], colorbar=True) #vmin=.05, vmax=.5,
plot_surf(fsaverage['infl_right'],join(gii_dir,f'{key}_align-ha_common_sub-mean_run-mean_hemi-rh.gii'), 
          title=f'{key} 4h', 
          hemi='right', view = 'lateral', cmap = 'inferno',axes=axes[1], colorbar=True) #vmin=.05, vmax=.5,

plt.show()
```
