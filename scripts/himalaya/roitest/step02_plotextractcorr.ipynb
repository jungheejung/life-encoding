```python
import numpy as np
import os, sys
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fname = 'split-r_pca-40_align-ha_common_sub-rid000005_run-1_roi-AIP_one_hemi-lh.npy'

# stack_dir = '/dartfs/rc/lab/D/DBIC/DBIC/f0042x1/life-encoding/results/himalaya/roitest_AIP_one/ha_common_pca-40'
stack_dir = '/Volumes/life-encoding/results/himalaya/roitest_AIP_one/ha_common_pca-40'
stack_data = {'bg': [], 'actions': [], 'agents': [], 'moten': []}
split_data = np.load(f'{stack_dir}/{fname}', allow_pickle=True).item()
for feature in stack_data:
    stack_data[feature].append(split_data[feature])
```

```python
stack_data
```

```python
stack_data['bg'][0].mean()

averages = {key: np.mean(np.array(values)) for key, values in stack_data.items()}

# Plot the average values
keys = list(averages.keys())
avg_values = list(averages.values())

plt.figure(figsize=(10, 6))
plt.bar(keys, avg_values, color='navy')
plt.xlabel('Keys')
plt.ylabel('Average Value')
plt.title('Average Values for Each Key')
sns.despine()
plt.show()
```
