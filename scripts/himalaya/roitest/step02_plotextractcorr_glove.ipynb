```python
import numpy as np
import os, sys
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
fname = 'split-r_pca-40_align-ha_common_sub-rid000005_run-1_roi-MT_one_hemi-lh.npy'

# stack_dir = '/dartfs/rc/lab/D/DBIC/DBIC/f0042x1/life-encoding/results/himalaya/roitest_AIP_one/ha_common_pca-40'
stack_dir = '/Volumes/life-encoding/results/himalaya/roitest_MT_one_glove/ha_common_pca-40'
stack_data = {'bg': [], 'actions': [], 'agents': [], 'moten': []}
split_data = np.load(f'{stack_dir}/{fname}', allow_pickle=True).item()
for feature in stack_data:
    stack_data[feature].append(split_data[feature])
```

```python
stack_data
```

```python
import pandas as pd

roi_df = {'performance': [],'feature space': []}
for k in stack_data:
    for v in stack_data[k][0][0]:
        roi_df['performance'].append(v)
        roi_df['feature space'].append(k)    
df = pd.DataFrame(roi_df)
# df.melt(var_name='key', value_name='value')
```

```python
df
```

```python
roiname = 'MT'
sns.barplot(x="feature space", 
              y="performance", 
              data=df,
              order=['bg', 'agents', 'actions', 'moten']
            ).set_title(f"datadir: {roiname}")
# plt.ylim(-0.025, 0.08)
sns.despine()
```
