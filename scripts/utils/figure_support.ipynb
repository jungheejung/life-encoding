```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
for key in ['Reds', 'Blues', 'Greens', 'Greys']:
# Define the custom red colormap
    reds_cmap = plt.get_cmap(key)

    # Generate a random 40x1400 matrix
    matrix = np.random.rand(10, 70) 

    # Plot the matrix using the custom colormap
    plt.figure(figsize=(7, 1))  # Adjust the figure size as needed
    plt.imshow(matrix, cmap=reds_cmap, aspect='auto')

    # Remove the axis for a cleaner look
    plt.axis('off')

    # Show the plot
    plt.show()
```
