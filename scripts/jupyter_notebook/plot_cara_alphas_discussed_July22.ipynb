This notebook plots the alpha values from Cara's original ridge regression analysis
The idea is to get a general idea of the alpha distribution and see if we can compare it with the alpha values in the tikreg analysis

# libraries

```python
import numpy as np
# import matplotlib; matplotlib.use('agg')
import os

import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```python
subject_list = ['sub-rid000001','sub-rid000005','sub-rid000006','sub-rid000009',
'sub-rid000012','sub-rid000014','sub-rid000017','sub-rid000019',
'sub-rid000024','sub-rid000027','sub-rid000031','sub-rid000032',
'sub-rid000033','sub-rid000034','sub-rid000036','sub-rid000037',
'sub-rid000038','sub-rid000041']
```

# functions

```python
def plot_alphas(sub_dir, hemi, title):
    data_dict = {} # empty dict
    data_list = [] # empty list
    for sub in subject_list:
        data_list.append(np.load(os.path.join(sub_dir,sub,hemi,'alphas.npy')))
    hist, bins = np.histogram(data_list, 50)
    _ = plt.hist(data_list, bins='auto')  # arguments are passed to np.histogram
    plt.title(title)
    # Text(0.5, 1.0, "Histogram with 'auto' bins")
    # x1 = np.random.normal(0, 0.8, 1000)
    # x2 = np.random.normal(-2, 1, 1000)
    # x3 = np.random.normal(3, 2, 1000)

    kwargs = dict(histtype='stepfilled', alpha=0.7, density=True, bins=10000, ec="k")

    plt.hist(data_list[0], **kwargs)
    plt.hist(data_list[1], **kwargs)
    plt.hist(data_list[2], **kwargs)
    plt.hist(data_list[3], **kwargs)
    plt.hist(data_list[4], **kwargs)
    plt.hist(data_list[5], **kwargs)
    plt.hist(data_list[6], **kwargs)
    plt.hist(data_list[7], **kwargs)
    plt.hist(data_list[8], **kwargs)
    plt.hist(data_list[9], **kwargs)
    plt.hist(data_list[10], **kwargs)
    plt.hist(data_list[11], **kwargs)
    plt.hist(data_list[12], **kwargs)
    plt.hist(data_list[13], **kwargs)
    plt.hist(data_list[14], **kwargs)
    plt.hist(data_list[15], **kwargs)
    plt.hist(data_list[16], **kwargs)
    plt.hist(data_list[17], **kwargs)

    plt.xlabel('alpha values')
    plt.ylabel('frequency')
    plt.show()
```
